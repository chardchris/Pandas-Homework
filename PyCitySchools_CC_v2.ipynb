```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "./Resources/schools_complete.csv"
student_data_to_load = "./Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
bins = [70,100]
bin_labels = ['Pass']
Math_PassCt = pd.cut(school_data_complete['math_score'],bins,labels=bin_labels,include_lowest=True)
Reading_PassCt = pd.cut(school_data_complete['reading_score'],bins,labels=bin_labels, include_lowest=True)

TotSchools = school_data_complete.school_name.nunique()
TotStudents = school_data_complete['Student ID'].nunique()
TotBudget = sum(school_data_complete.budget.unique())
AvgMath = school_data_complete.math_score.mean()
AvgRead = school_data_complete.reading_score.mean()
school_data_complete['Total_Score']= (school_data_complete['math_score']+school_data_complete['reading_score'])/2
AvgTot = school_data_complete.Total_Score.mean()
MathPass = Math_PassCt.count()/school_data_complete.math_score.count()  
ReadPass = Reading_PassCt.count()/school_data_complete.reading_score.count()
Dist_Summ_df = pd.DataFrame({'Total Schools':[TotSchools],
                              'Total Students':[TotStudents],
                              'Total Budget':[TotBudget],
                              'Average Math Score':[AvgMath] ,
                              'Average Reading Score':[AvgRead] ,
                              'Passing Rate':[AvgTot] ,
                              '% Passing Math':[MathPass] ,
                              '% Passing Reading':[ReadPass]})
Dist_Summ_df
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
School_Grp = school_data_complete.groupby(['school_name'])
School_Math_Pass = school_data_complete[school_data_complete.math_score >= 70]
Math_Pass_Grp = School_Math_Pass.groupby(['school_name'])
School_Read_Pass = school_data_complete[school_data_complete.reading_score >=70]
Read_Pass_Grp = School_Read_Pass.groupby(['school_name'])
School_1 = School_Grp['type','budget'].first()
School_1['Total Students'] = School_Grp.student_name.count()
School_1['Per Student Budget'] = School_1['budget']/School_1['Total Students']
School_1['Average Math Score'] = School_Grp.math_score.mean()
School_1['Average Reading Score'] = School_Grp.reading_score.mean()
School_1['# Passing Math'] = Math_Pass_Grp.student_name.count()
School_1['# Passing Reading'] = Read_Pass_Grp.student_name.count()
School_1['% Passing Reading'] = (School_1['# Passing Reading']/School_1['Total Students'])*100 
School_1['% Passing Math'] = (School_1['# Passing Math']/School_1['Total Students'])*100
School_1['Overall Passing Rate'] = (School_1['% Passing Math'] + School_1['% Passing Reading']) /2

School_2 = School_1.rename(columns={'type':'School Type','budget':'Total Budget'})
School_3 = School_2.drop(columns = ['# Passing Math','# Passing Reading'])
```

```python
Top_Schools = School_3.sort_values('Overall Passing Rate',ascending=False)

Top_Schools.head()
```
